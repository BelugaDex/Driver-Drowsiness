```python
images_path=os.path.join('data','images') #data/images
labels = ['awake','drowsy']
number_imgs= 20

cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(images_path, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import pygame  # Import pygame for audio
import cv2
import numpy as np
import dlib
from imutils import face_utils
import threading

# Initialize pygame mixer for sound
pygame.mixer.init()
pygame.mixer.music.load("alarm.wav")  # Load alarm sound

# Initializing the camera
cap = cv2.VideoCapture(0)

# Initializing the face detector and landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Status variables
sleep = 0
drowsy = 0
active = 0
status = ""
color = (0, 0, 0)
alarm_playing = False

def compute(ptA, ptB):
    return np.linalg.norm(ptA - ptB)

def blinked(a, b, c, d, e, f):
    up = compute(b, d) + compute(c, e)
    down = compute(a, f)
    ratio = up / (2.0 * down)
    
    if ratio > 0.25:
        return 2
    elif 0.21 <= ratio <= 0.25:
        return 1
    else:
        return 0

def play_alarm():
    """Plays an alarm sound (runs in a separate thread)."""
    global alarm_playing
    if not alarm_playing:
        alarm_playing = True
        pygame.mixer.music.play()
        while pygame.mixer.music.get_busy():  # Wait until sound finishes
            continue
        alarm_playing = False  # Reset after playing

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    eye_frame = np.zeros_like(frame)  # Blank frame for eye tracking
    landmarks_frame = frame.copy()  # Copy frame for landmark visualization

    for face in faces:
        landmarks = predictor(gray, face)
        landmarks = face_utils.shape_to_np(landmarks)

        left_blink = blinked(landmarks[36], landmarks[37], landmarks[38], 
                             landmarks[41], landmarks[40], landmarks[39])
        right_blink = blinked(landmarks[42], landmarks[43], landmarks[44], 
                              landmarks[47], landmarks[46], landmarks[45])

        # Extract left and right eye regions
        left_eye_pts = landmarks[36:42]
        right_eye_pts = landmarks[42:48]

        # Draw contours around eyes
        cv2.polylines(frame, [left_eye_pts], isClosed=True, color=(255, 255, 0), thickness=1)
        cv2.polylines(frame, [right_eye_pts], isClosed=True, color=(255, 255, 0), thickness=1)

        # Crop and resize the eyes
        x1, y1 = np.min(left_eye_pts, axis=0)
        x2, y2 = np.max(left_eye_pts, axis=0)
        left_eye_roi = gray[y1:y2, x1:x2]

        x1, y1 = np.min(right_eye_pts, axis=0)
        x2, y2 = np.max(right_eye_pts, axis=0)
        right_eye_roi = gray[y1:y2, x1:x2]

        # Resize for display
        left_eye_display = cv2.resize(left_eye_roi, (150, 80))
        right_eye_display = cv2.resize(right_eye_roi, (150, 80))

        # Place in eye tracking window
        eye_frame[50:130, 50:200] = cv2.cvtColor(left_eye_display, cv2.COLOR_GRAY2BGR)
        eye_frame[50:130, 250:400] = cv2.cvtColor(right_eye_display, cv2.COLOR_GRAY2BGR)

        # Draw white dots for all 68 landmarks
        for (x, y) in landmarks:
            cv2.circle(landmarks_frame, (x, y), 2, (255, 255, 255), -1)  # White dots

        if left_blink == 0 or right_blink == 0:
            sleep += 1
            drowsy = 0
            active = 0
            if sleep > 6:
                status = "SLEEPING !!!"
                color = (255, 0, 0)

                if not alarm_playing:
                    threading.Thread(target=play_alarm, daemon=True).start()

        elif left_blink == 1 or right_blink == 1:
            sleep = 0
            active = 0
            drowsy += 1
            if drowsy > 6:
                status = "Drowsy !"
                color = (0, 0, 255)

                if not alarm_playing:
                    threading.Thread(target=play_alarm, daemon=True).start()

        else:
            drowsy = 0
            sleep = 0
            active += 1
            if active > 6:
                status = "Active :)"
                color = (0, 255, 0)

        cv2.putText(frame, status, (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3)

    # Show main frame
    cv2.imshow("Frame", frame)
    cv2.imshow("Eye Tracking", eye_frame)  # Show eye tracking window
    cv2.imshow("Face Landmarks", landmarks_frame)  # Show face with white dots

    if cv2.waitKey(1) == 27:  # Press 'Esc' to exit
        break

cap.release()
cv2.destroyAllWindows()
pygame.mixer.quit()  # Clean up pygame
```

```python
import pygame
import cv2
import numpy as np
import dlib
from imutils import face_utils
import threading
import pandas as pd
import matplotlib.pyplot as plt
import time

# -------------------- Time Input Function -------------------- #
def parse_duration():
    raw_input = input("Enter duration (in format MM:SS or SS): ")
    try:
        if ":" in raw_input:
            minutes, seconds = map(int, raw_input.strip().split(":"))
            return minutes * 60 + seconds
        else:
            return float(raw_input.strip())
    except Exception as e:
        print(f"Invalid input: {e}. Defaulting to 30 seconds.")
        return 30.0

# Get user-defined duration
duration = parse_duration()

# -------------------- Audio Alert -------------------- #
pygame.mixer.init()
pygame.mixer.music.load("alarm.wav")

# -------------------- Video Capture -------------------- #
cap = cv2.VideoCapture(0)
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0 or fps is None or np.isnan(fps):
    fps = 30  # default fallback

# -------------------- Dlib Models -------------------- #
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# -------------------- Variables -------------------- #
sleep = drowsy = active = 0
status = ""
color = (0, 0, 0)
alarm_playing = False

wakefulness_level = 1.0
blink_counter = 0
blink_threshold = 5

ear_data = []
frame_count = 0
start_time = time.time()

# -------------------- Helper Functions -------------------- #
def compute(ptA, ptB):
    return np.linalg.norm(ptA - ptB)

def eye_aspect_ratio(a, b, c, d, e, f):
    up = compute(b, d) + compute(c, e)
    down = compute(a, f)
    return up / (2.0 * down)

def play_alarm():
    global alarm_playing
    if not alarm_playing:
        alarm_playing = True
        pygame.mixer.music.play()
        while pygame.mixer.music.get_busy():
            continue
        alarm_playing = False

# -------------------- Main Loop -------------------- #
while True:
    if time.time() - start_time > duration:
        break

    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    landmarks_frame = frame.copy()

    for face in faces:
        landmarks = predictor(gray, face)
        landmarks = face_utils.shape_to_np(landmarks)

        left_ear = eye_aspect_ratio(landmarks[36], landmarks[37], landmarks[38],
                                    landmarks[41], landmarks[40], landmarks[39])
        right_ear = eye_aspect_ratio(landmarks[42], landmarks[43], landmarks[44],
                                     landmarks[47], landmarks[46], landmarks[45])
        avg_ear = (left_ear + right_ear) / 2.0

        # Draw eye contours
        cv2.polylines(frame, [landmarks[36:42]], isClosed=True, color=(255, 255, 0), thickness=1)
        cv2.polylines(frame, [landmarks[42:48]], isClosed=True, color=(255, 255, 0), thickness=1)

        for (x, y) in landmarks:
            cv2.circle(landmarks_frame, (x, y), 2, (255, 255, 255), -1)

        # Blink Detection
        if avg_ear < 0.18:
            blink_counter += 1
        else:
            if 2 <= blink_counter <= blink_threshold:
                print("Blink detected!")
            blink_counter = 0

        # Drowsiness Detection
        if avg_ear < 0.18:
            sleep += 1
            drowsy = 0
            active = 0
            wakefulness_level = max(0.0, wakefulness_level - 0.15)
            if sleep > 3:
                status = "SLEEPING !!!"
                color = (255, 0, 0)
                if not alarm_playing:
                    threading.Thread(target=play_alarm, daemon=True).start()

        elif 0.18 <= avg_ear < 0.23:
            sleep = 0
            active = 0
            drowsy += 1
            wakefulness_level = max(0.3, wakefulness_level - 0.07)
            if drowsy > 2:
                status = "Drowsy !"
                color = (0, 0, 255)
                if not alarm_playing:
                    threading.Thread(target=play_alarm, daemon=True).start()

        else:
            drowsy = 0
            sleep = 0
            active += 1
            wakefulness_level = min(1.0, wakefulness_level + 0.1)
            if active > 5:
                status = "Active :)"
                color = (0, 255, 0)

        # Store EAR Data
        frame_count += 1
        timestamp = frame_count / fps
        ear_data.append({
            'Frame': frame_count,
            'Time (s)': timestamp,
            'EAR': avg_ear,
            'Wakefulness': wakefulness_level,
            'Drowsiness': 1.0 - wakefulness_level
        })

        # Display info
        cv2.putText(frame, status, (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3)
        cv2.putText(frame, f"Wakefulness: {wakefulness_level:.2f}", (100, 150),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)

    cv2.imshow("Frame", frame)
    cv2.imshow("Face Landmarks", landmarks_frame)

    if cv2.waitKey(1) == 27:  # ESC
        break

# -------------------- Cleanup -------------------- #
cap.release()
cv2.destroyAllWindows()
pygame.mixer.quit()

# -------------------- Save Data -------------------- #
df = pd.DataFrame(ear_data)
df.to_excel("ear_wakefulness_data.xlsx", index=False)

# -------------------- Plotting -------------------- #
# EAR Plot
plt.figure(figsize=(12, 5))
plt.plot(df['Time (s)'], df['EAR'], label="EAR", color='blue')
plt.axhline(0.18, color='red', linestyle='--', label='Sleep Threshold')
plt.axhline(0.23, color='orange', linestyle='--', label='Drowsy Threshold')
plt.xlabel("Time (seconds)")
plt.ylabel("EAR")
plt.title("Eye Aspect Ratio (EAR) Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("ear_plot.png")
plt.show()

# Wakefulness Plot
plt.figure(figsize=(12, 5))
plt.plot(df['Time (s)'], df['Wakefulness'], label="Wakefulness", color='green')
plt.xlabel("Time (seconds)")
plt.ylabel("Wakefulness")
plt.title("Wakefulness Over Time")
plt.grid(True)
plt.tight_layout()
plt.savefig("wakefulness_plot.png")
plt.show()

# Drowsiness Plot
plt.figure(figsize=(12, 5))
plt.plot(df['Time (s)'], df['Drowsiness'], label="Drowsiness Level", color='purple')
plt.xlabel("Time (seconds)")
plt.ylabel("Drowsiness Level (1 - Wakefulness)")
plt.title("Drowsiness Level Over Time")
plt.grid(True)
plt.tight_layout()
plt.savefig("drowsiness_plot.png")
plt.show()

# -------------------- Area Under Curve -------------------- #
area_under_drowsiness = np.trapz(df['Drowsiness'], df['Time (s)'])
area_per_time = area_under_drowsiness / duration
print(f"Area under Drowsiness Curve: {area_under_drowsiness:.4f}")
print(f"Area/Time Ratio: {area_per_time:.4f}")
```
